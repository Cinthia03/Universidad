```python
# -------------------------------------------
# INSTALACIÓN DE LIBRERÍAS
# -------------------------------------------
!pip install spacy
!python -m spacy download es_core_news_md
!pip install nltk

# -------------------------------------------
# IMPORTACIÓN DE LIBRERÍAS
# -------------------------------------------
import spacy
import nltk
from nltk.corpus import wordnet as wn

nltk.download('omw-1.4')
nltk.download('wordnet')

# Cargar modelo de spaCy
nlp = spacy.load("es_core_news_md")

# -------------------------------------------
# FUNCIÓN DE ANÁLISIS SEMÁNTICO
# -------------------------------------------
def analisis_semantico(oracion):

    print("\n============================")
    print("🔍 ANÁLISIS COMPLETO DE LA ORACIÓN")
    print("============================\n")

    doc = nlp(oracion)

    # --- Tokenización
    print("📌 TOKENS:")
    for token in doc:
        print(f"  - {token.text}")

    # --- Etiquetado morfosintáctico
    print("\n📌 ETIQUETAS GRAMATICALES:")
    for token in doc:
        print(f"{token.text:15} → {token.pos_} ({token.tag_})")

    # --- Dependencias sintácticas
    print("\n📌 DEPENDENCIAS:")
    for token in doc:
        print(f"{token.text:15} → jefe: {token.head.text},  relación: {token.dep_}")

    # --- Entidades nombradas
    print("\n📌 ENTIDADES RECONOCIDAS:")
    if doc.ents:
        for ent in doc.ents:
            print(f"{ent.text:20} → {ent.label_}")
    else:
        print("No se encontraron entidades.")

    # --- Análisis semántico con WordNet
    print("\n📌 ANÁLISIS SEMÁNTICO (WordNet):")

    for token in doc:
        syns = wn.synsets(token.text, lang='spa')
        if syns:
            print(f"\n🔹 Palabra: {token.text}")
            print(f"   - Significado: {syns[0].definition()}")
            print(f"   - Sinónimos: {[l.name() for l in syns[0].lemmas()]}")
        else:
            print(f"\n🔹 Palabra: {token.text}")
            print("   - No se encontraron significados en WordNet.")

    # --- Similaridad semántica entre palabras importantes
    print("\n📌 SIMILARIDAD SEMÁNTICA ENTRE PALABRAS:")
    contenido = [t for t in doc if t.pos_ in ("NOUN", "VERB", "ADJ")]

    for i in range(len(contenido)):
        for j in range(i+1, len(contenido)):
            sim = contenido[i].similarity(contenido[j])
            print(f"{contenido[i].text} ↔ {contenido[j].text} = {sim:.3f}")

# -------------------------------------------
# PROGRAMA PRINCIPAL
# -------------------------------------------
texto = input("Ingrese una oración para analizar: ")
analisis_semantico(texto)

```
